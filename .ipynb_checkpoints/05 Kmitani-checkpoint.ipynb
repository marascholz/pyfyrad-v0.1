# Kmitání
Mezi nejvýznamnější typ pohybu patří kmitavý pohyb, jehož exemplárním příkladem je kmitání závaží zavěšeného na pružině. Po zavěšení se závaží ustálí v rovnovážné poloze, kde prodloužení pružiny kompenzuje tíhovou sílu. Tíhovou sílu dále tedy nemusíme uvažovat. Při vychýlení z této rovnovážné polohy působí na závaží 'vratná' elastická síla od pružiny $F_p = - kx$, kde $k$ je tuhost pružiny a $x$ velikost výchylky z rovnovážné polohy. Pokud na závaží nepůsobí žádné třecí a odporové síly, pak lze na základě II. Newtonova zákona psát pohybovou rovnici
$$
F = ma = -kx
$$
Pro zrychlení tak máme
$$
a = \frac{k}{m}x
$$
pro rychlost pak platí $\Delta v = a\cdot \Delta t$.
Pro časový vývoj polohy pak $\Delta x = v\cdot \Delta t$ a časový vývoj rychlostí $\Delta v = a\cdot \Delta t$.

```python
import numpy as np
import matplotlib.pyplot as plt

g = 9.8 # m.s-2
k = 100 # N/m
m = 1 # [kg]
x = 0.1 # [m], pocatecni vychylka 5 cm
v = 0 # [m/s], pocatecni rychlost nulova
dt = 0.001
t = np.arange(0,5,dt) # tmin,tmax,dt
R = []
for tt in t:
    a = g - k/m*x
    v = v + a*dt
    x = x + v*dt
    R.append([x,v,a])
R = np.array(R)
plt.plot(t,R[:,0],'r-',label='x [m]')    
plt.plot(t,R[:,1],'g-',label='v [m/s]')    
plt.plot(t,R[:,2],'b-',label='a [m/s2]')
plt.legend()
plt.grid()
```

```python
from scipy.integrate import odeint
g = 9.8 # m.s-2
k = 100 # N/m
m = 1 # [kg]
x0 = 0.1 # [m], pocatecni vychylka 5 cm
v0 = 0.0 # [m/s], pocatecni rychlost nulova
t = np.arange(0,5,0.01)
def rovnice_pohybu(xv,t,m,k,g):
    x = xv[0]
    v = xv[1]
    dxdt = v
    dvdt = -k/m*x
    return np.array([dxdt,dvdt])
R = odeint(rovnice_pohybu,[x0,v0],t,args=(m,k,g))
plt.plot(t,R[:,0],label='x')
plt.plot(t,R[:,1],label='v')
plt.legend()

```

# Kmitání s tlumením
Představme si ale, že závaží na pružině je třeba umístěno ve vodě, kde na něj působí odporová síla.
Velikost odporové síly je při malých rychlostech přímo úměrná velikosti rychlosti, tedy $F_o = -pv$, kde koeficient $p$ souvisí s viskozitou prostředí.
Pohybová rovnice se pak mění na
$$ ma = -kx - pv$$

```python
## Zde je překopírován předchozí kód. Modifukujte ho tak, aby byl zahrnut odpor prostředí.
## Potom měňte parametr p a sledujte, jak se mění průběh x(t)
from scipy.integrate import odeint
g = 9.8 # m.s-2
k = 100 # N/m
m = 1 # [kg]
x0 = 0.1 # [m], pocatecni vychylka 5 cm
v0 = 0.0 # [m/s], pocatecni rychlost nulova
t = np.arange(0,5,0.01)
def rovnice_pohybu(xv,t,m,k,g):
    x = xv[0]
    v = xv[1]
    dxdt = v
    dvdt = -k/m*x
    return np.array([dxdt,dvdt])
R = odeint(rovnice_pohybu,[x0,v0],t,args=(m,k,g))
plt.plot(t,R[:,0],label='x')
plt.plot(t,R[:,1],label='v')
plt.legend()
```

# řešení

```python
from scipy.integrate import odeint
g = 9.8 # m.s-2
k = 100 # N/m
m = 1 # [kg]
pp = [0.1,1,5,10,15,20]
x0 = 0.1 # [m], pocatecni vychylka 5 cm
v0 = 0.0 # [m/s], pocatecni rychlost nulova
t = np.arange(0,2,0.01)
def rovnice_pohybu(xv,t,m,k,g,p):
    x = xv[0]
    v = xv[1]
    dxdt = v
    dvdt = (-k*x - p*v)/m
    return np.array([dxdt,dvdt])
for i,p in enumerate(pp):
    R = odeint(rovnice_pohybu,[x0,v0],t,args=(m,k,g,p))
    plt.plot(t,R[:,0],label='p='+str(p))
plt.legend()
```

# Buzení
Mějme nyní závaží na pružině, které je v odporujícím prostředí (nebo podléhá vnitřnímu tření v pružině).
K tomu však ještě budeme pružinu držet v ruce a slabě s ní pohupovat a tím periodicky měnit sílu, kterou pružina působí na závaží.
K dříve zavedeným silám tedy ještě přidáme vnější sílu o velikosti $F = bsin(\omega t)$, která se periodicky mění s frekvencí $\omega$.
Vlastní frekvence kmitání netlumeného a nebuzeného oscilátoru je přitom $\omega_0 = \sqrt{k/m}$.
Tušíme, že pokud bude budící frekvence $\omega$ blízká vlastní frekvenci $\omega_0$, bude závaží dobře 'absorbovat' budící energii a výchylka se bude zvětšovat.

```python
from scipy.integrate import odeint
g = 9.8 # m.s-2
k = 100 # N/m
m = 1 # [kg]
print('omega0',np.sqrt(k/m))
p = 0.1
b = -0.5
omega = 10 #rad/s
x0 = 0.1 # [m], pocatecni vychylka 5 cm
v0 = 0.0 # [m/s], pocatecni rychlost nulova
t = np.arange(0,20,0.01)
def rovnice_pohybu(xv,t,m,k,g,p,b,omega):
    x = xv[0]
    v = xv[1]
    dxdt = v
    dvdt = (-k*x - p*v + b*np.sin(omega*t))/m
    return np.array([dxdt,dvdt])
R = odeint(rovnice_pohybu,[x0,v0],t,args=(m,k,g,p,b,omega))
plt.plot(t,R[:,0],label='p='+str(p))
plt.legend()
```

```python
from scipy.integrate import odeint
g = 9.8 # m.s-2
k = 100 # N/m
m = 1 # [kg]
print('omega0',np.sqrt(k/m))
p = 0.1
b = -0.2
omegas = np.arange(2,30,1) #rad/s
A = np.zeros(len(omegas))
x0 = 0.1 # [m], pocatecni vychylka 5 cm
v0 = 0.0 # [m/s], pocatecni rychlost nulova
t = np.arange(0,50,0.01)
def rovnice_pohybu(xv,t,m,k,g,p,b,omega):
    x = xv[0]
    v = xv[1]
    dxdt = v
    dvdt = (-k*x - p*v + b*np.sin(omega*t))/m
    return np.array([dxdt,dvdt])
for i,omega in enumerate(omegas):
    R = odeint(rovnice_pohybu,[x0,v0],t,args=(m,k,g,p,b,omega))
    A = np.max(R[-int(len(t)*0.1):-1,0])
plt.plot(omegas,A)
```
